```python
#Importing the necessary libraries
import numpy as np
import pandas as pd
import glob
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
w = glob.glob(r"data\*.nc")
#print(w)
# from
#https://cds.climate.copernicus.eu/cdsapp#!/dataset/reanalysis-era5-single-levels?tab=overview
```

```python
import xarray as xr
WindSpeed_data = xr.open_mfdataset(w)
```

```python
print(WindSpeed_data)
```

```python
import rioxarray

WindSpeed_data.rio.set_spatial_dims(x_dim="longitude", y_dim="latitude", inplace=True)
WindSpeed_data.rio.write_crs("epsg:4326", inplace=True)
```

```python
import geopandas
G5_Sahel = geopandas.read_file("data\G5_Sahel_adm2N\G5_Sahel_adm2N.shp", crs="epsg:4326")
```

```python
G5_Sahel.head()
```

```python
G5_Sahel.drop(['ADMIN0NAME','ADMIN0PCOD','ADMIN1NAME','ADMIN1PCOD','INITIAL','Shape_Leng','Shape_Area'], axis=1, inplace=True)
```

```python
G5_Sahel.head()
```

```python
clipped = WindSpeed_data.rio.clip(G5_Sahel.geometry, G5_Sahel.crs, drop=True)
```

```python
import matplotlib.pyplot as plt
ax=G5_Sahel.plot(facecolor='none',edgecolor='black')
sahel=clipped.v10.isel(time=0).plot(ax=ax,zorder=-1)
```

```python
import matplotlib.pyplot as plt
ax=G5_Sahel.plot(facecolor='none',edgecolor='black')
sahel=clipped.u10.isel(time=0).plot(ax=ax,zorder=-1)
```

```python
# filter Comoe
Comoe = G5_Sahel.loc[G5_Sahel['ADMIN2NAME'] == 'Comoe']
Comoe.head()
```

```python
ax=Comoe.plot(facecolor='none',edgecolor='black')
#clipped.u10.isel(time=0).plot(ax=ax,zorder=-1)
clipped = WindSpeed_data.rio.clip(Comoe.geometry, Comoe.crs, drop=True)
clipped.u10.isel(time=0).plot(ax=ax,zorder=-1)
```

```python
ax=Comoe.plot(facecolor='none',edgecolor='black')
clipped.v10.isel(time=0).plot(ax=ax,zorder=-1)
```

```python
clipped
```

```python
clipped['wind_speed'] = clipped['u10']**2 + clipped['v10']**2
clipped['wind_speed'] = np.sqrt(clipped['wind_speed'])
```

```python
clipped
```

```python
ax=Comoe.plot(facecolor='none',edgecolor='black')
clipped.wind_speed.isel(time=0).plot(ax=ax,zorder=-1)
```

```python
ax=Comoe.plot(facecolor='none',edgecolor='black')
clipped.wind_speed.isel(time=-1).plot(ax=ax,zorder=-1)
```

```python
#pip install xagg
import xagg as xa
```

```python
# Get overlap between pixels and polygons
weightmap = xa.pixel_overlaps(clipped, Comoe, subset_bbox=False)
```

```python
# gives the weighted average of the values in raster based on each pixel’s relative area overlap with the polygons.
aggregated = xa.aggregate(clipped, weightmap)
```

```python
aggregated
```

```python
ds = aggregated.to_dataset()
```

```python
ds.head()
```

```python
# Example as a pandas dataframe
df_out = aggregated.to_dataframe()
df_out.head()
```

```python
ds.to_dataframe().to_csv("2023_Comoe_Wind_Speed_10m_hours.csv")
```
