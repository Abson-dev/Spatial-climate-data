```python
#Importing the necessary libraries
import numpy as np
import glob
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
w = glob.glob(r"data\2019_WindSpeed\*.nc")
#print(w)
```

```python
import xarray as xr
WindSpeed_data = xr.open_mfdataset(w)
```

```python
print(WindSpeed_data)
```

```python
WindSpeed_data.Wind_Speed_10m_Mean.isel(time=0).plot()
```

```python
import rioxarray

WindSpeed_data.rio.set_spatial_dims(x_dim="lon", y_dim="lat", inplace=True)
WindSpeed_data.rio.write_crs("epsg:4326", inplace=True)
```

```python
import geopandas
G5_Sahel = geopandas.read_file("data\G5_Sahel_adm2N\G5_Sahel_adm2N.shp", crs="epsg:4326")
```

```python
G5_Sahel.head()
```

```python
clipped = WindSpeed_data.rio.clip(G5_Sahel.geometry, G5_Sahel.crs, drop=True)
```

```python
import matplotlib.pyplot as plt
ax=G5_Sahel.plot(facecolor='none',edgecolor='black')
sahel=clipped.Wind_Speed_10m_Mean.isel(time=0).plot(ax=ax,zorder=-1)
```

```python
#pip install xagg
import xagg as xa
```

```python
# Get overlap between pixels and polygons
weightmap = xa.pixel_overlaps(WindSpeed_data, G5_Sahel, subset_bbox=False)
```

```python
aggregated = xa.aggregate(WindSpeed_data, weightmap)
```

```python
ds = aggregated.to_dataset()
# Drop columns multiple column 
```

```python
ds.head()
```

```python
# Example as a pandas dataframe
df_out = aggregated.to_dataframe()
df_out.head()
```
